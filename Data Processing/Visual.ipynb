```python
import pandas as pd
import numpy as np
import datetime
from matplotlib.ticker import FuncFormatter
import matplotlib.pyplot as plt
import warnings
import math as ma
warnings.filterwarnings("ignore")
```

```python
df1 = pd.read_csv('movie_metadata.csv')
df1.head()
```

```python
df_count = df1.groupby(["title_year", "country"]).agg(
    count_col=pd.NamedAgg(column="movie_title", aggfunc="count"))

df_count.head(30)
```

```python
df_count1 = df1.groupby(["country"]).agg(
    count_col=pd.NamedAgg(column="movie_title", aggfunc="count"))

df_count1.head(30)
```

```python
df_count1.to_csv('map1.csv')
```

```python
df_count3 = df1.groupby('title_year', as_index=False)['gross'].sum()
df_count3.tail(30)
```

```python
df_count3.to_csv('Gross.csv')
```

```python
df_count4 = df1.groupby(["genres", "country"]).agg(
    count_col=pd.NamedAgg(column="movie_title", aggfunc="count"))

df_count4.tail(30)
```

```python
df_count7 = df1.groupby(["genres"]).agg(
    count_col=pd.NamedAgg(column="movie_title", aggfunc="count"))

df_count7.tail(30)
```

```python
df_newww=df1.groupby('title_year', as_index=False)['budget'].sum()
df_newww.tail(30)
```

```python
df_row = pd.concat([df_count3, df_newww])
df_row.tail(30)
```

```python
df_newww.to_csv('Budget.csv')
```
